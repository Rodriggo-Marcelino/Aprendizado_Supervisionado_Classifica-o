```python
import pandas as pd 
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "iframe"
```

```python
dataset = pd.read_csv('heart.csv', sep=',', encoding = 'iso-8859-1')
```

```python
dataset.head(10)
```

```python
dataset.shape
```

# Analise das variaveis 

## Age = idade 

```python
dataset['Age'].value_counts()
```

```python
dataset['Age'].value_counts().sort_index()
```

```python
dataset.columns
```

```python
hist1 = px.histogram(dataset, x = "Age", nbins=60)
hist1.update_layout(width=800, height=500,title_text='Distribuição das idades')
hist1.show()
```

```python
sns.histplot(dataset, x='Age', bins = 30, color = 'orange', kde = True, stat='count');
```

# Sex = sexo

```python
dataset['Sex'].value_counts()
```

```python
sns.countplot(x = 'Sex', data = dataset);
```

# Chest Pain Type = tipo de dor no peito

```python
dataset['ChestPainType'].value_counts()
```

- ASY = assintomatico 
- NAP = dor não anginosa
- ATA = angina atipica 
- TA = angina tipica 
---
Agina = dor no peito causada pela diminuição do fluxo de sangue no coração 

```python
sns.countplot(x = 'ChestPainType', data = dataset);
```

# Resting BP (blood pressure) = pressão sanguinea em repouso em mmHG(sistólica)

```python
dataset.RestingBP.value_counts().sort_index()
```

```python
sns.histplot(dataset, x = 'RestingBP', bins=30, color = 'orange', kde= True, stat='count');
```

# Cholesterol = colesterol sérico (mg/dl)

```python
dataset.Cholesterol.value_counts().sort_index()
```

```python
sns.histplot(dataset, x = 'Cholesterol', bins=30, color='orange', kde = True, stat='count');
```

# Fasting BS (Blood sugar) = açucar no sangue em jejum (mg/dl)

```python
dataset.FastingBS.value_counts().sort_index()
```

- 0 fasting BS < 120 mg/dll (não diabetico)
- 1 fasting BS >= 120 mg/dll (diabetico)

```python
sns.countplot(x='FastingBS', data=dataset);
```
